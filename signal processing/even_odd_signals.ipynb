## Even and Odd Signals

```python
import numpy as np
import matplotlib.pyplot as plt


def signal(t):
    return np.where(np.abs(t) < 1, 2 - np.abs(t), 0)
  

def plot_signal(t, s):
    plt.figure(figsize=(10, 8))
    plt.subplot(2, 1, 1)
    plt.plot(t, s)
    plt.xlabel('t')
    plt.ylabel('s(t)')
    plt.title('Signal')
    plt.grid(True)
    plt.legend(['s(t)'])
    plt.show()
    
# Time vector
t = np.linspace(-2, 3, 400)
# Signal
s = signal(t)

plot_signal(t, s)
  
```

```python
# Return Even part of the signal
def even_part(signal, t):
    return 0.5 * (signal(t) + signal(-t))
  
  
# Return Odd part of the signal
def odd_part(signal, t):
    return 0.5 * (signal(t) - signal(-t))
  
  
# Even part of the signal
s_even = even_part(signal, t)

# Odd part of the signal
s_odd = odd_part(signal, t)


plot_signal(t, s_even)
```

```python
plot_signal(t, s_odd)
```

```python
# Unit step signal
def unit_step(t):
    return np.where(t >= 0, 1, 0)
  
  
t = np.linspace(-2, 3, 400)
s = unit_step(t)
plot_signal(t, s)
```

```python
# Plot Even part of the unit step signal
s_even = even_part(unit_step, t)
plot_signal(t, s_even)
```

```python
# Plot Odd part of the unit step signal
s_odd = odd_part(unit_step, t)
plot_signal(t, s_odd)
```

```python
# Rectangular signal
def rectangular(t):
    return np.where(np.logical_and(t >= -0.5, t <= 0.5),  1, 0)
  
t = np.linspace(-2, 3, 400)
s = rectangular(t)
plot_signal(t, s)
```

```python
# Plot Even part of the rectangular signal
s_even = even_part(rectangular, t)
plot_signal(t, s_even)

# Plot Odd part of the rectangular signal
s_odd = odd_part(rectangular, t)
plot_signal(t, s_odd)
```

```python
# Exponential Decay Signal
def exp_decay(t):
    return np.exp(-t) * (t >= 0)

t = np.linspace(-2, 3, 400)
s = exp_decay(t)

plot_signal(t, s)
```

```python
# Plot Even part of the exponential decay signal
s_even = even_part(exp_decay, t)
plot_signal(t, s_even)

# Plot Odd part of the exponential decay signal
s_odd = odd_part(exp_decay, t)
plot_signal(t, s_odd)
```

```python
# Sawtooth signal
def sawtooth(t):
    return t - np.floor(t)
  
t = np.linspace(-2, 3, 400)
s = sawtooth(t)
plot_signal(t, s)
```

```python
# Plot Even part of the sawtooth signal
s_even = even_part(sawtooth, t)
plot_signal(t, s_even)

# Plot Odd part of the sawtooth signal
s_odd = odd_part(sawtooth, t)
plot_signal(t, s_odd)
```

```python
# Gaussian signal
def gaussian(t):
    return np.exp(-t**2)
  
t = np.linspace(-2, 3, 400)
s = gaussian(t)
plot_signal(t, s)
```

```python
# Plot Even part of the gaussian signal
s_even = even_part(gaussian, t)
plot_signal(t, s_even)

# Plot Odd part of the gaussian signal
s_odd = odd_part(gaussian, t)
plot_signal(t, s_odd)
```
