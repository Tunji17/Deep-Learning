## Basic Signal Operations

```python
import numpy as np
import matplotlib.pyplot as plt



def signal(t):
    # s(t) = t+1, -1 <= t <= 0
    # s(t) = 1-t/2, 0 < t <= 2
    # s(t) = 0, otherwise   
    return np.where((t >= -1) & (t <= 0), t+1, np.where((t > 0) & (t <= 2), 1-t/2, 0))
  
  
def plot_signal(t, s):
    plt.figure(figsize=(10, 8))
    plt.subplot(2, 1, 1)
    plt.plot(t, s)
    plt.xlabel('t')
    plt.ylabel('s(t)')
    plt.title('Signal')
    plt.grid(True)
    plt.show()
    
# Time vector
t = np.linspace(-2, 3, 400)
# Signal
s = signal(t)

plot_signal(t, s)
```

```python
# Time reversal and scaling x(-t/2)
def time_revers_and_scale(t):
    # Scaling by 1/2 this expands the signal by a factor of 2 in the time domain
    # Reversing the signal in time domain
    return signal(-t/2)

t = np.linspace(-5, 3, 400)
s = time_revers_and_scale(t)

plot_signal(t, s)
```

```python

# Time reversal and shift x(2(t+2))
def time_scale_and_shift(t):
    # Scaling by 2 this compresses the signal by a factor of 2 in the time domain
    # Shifting the signal in time domain by 2
    return signal(2*(t+2))
  
t = np.linspace(-4, 2, 400)
s = time_scale_and_shift(t)

plot_signal(t, s)
```

```python
# Time reversal and shift x(-(t+1))
def time_reverse_and_shift(t):
    # Reversing the signal in time domain
    # Shifting the signal in time domain by 2
    return signal(-(t+1))
  
t = np.linspace(-4, 1, 400)
s = time_reverse_and_shift(t)

plot_signal(t, s)
```
